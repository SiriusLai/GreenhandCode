```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
B1=np.arange(6,-7,-1)
B1
```

```python
V1=np.array([1.5,1.266,1.020,0.773,0.517,0.256,0,-0.255,-0.511,-0.765,-1.006,-1.243,-1.463])
V1
```

```python
plt.plot(B1,V1,'o',label='Experiment Data')
k1,k2=np.polyfit(B1,V1,1)
xs=np.linspace(-7,7,100)
ys=k1*xs+k2
plt.plot(xs,ys,label='Least Square Method Fit')
plt.plot(xs,0*xs,'green',0*ys,ys,'green')
plt.xlabel('Magnetic Induction/Gs',fontsize=14)
plt.ylabel('Output Voltage/V',fontsize=14)
plt.legend(frameon=False)
```

```python
angle=np.arange(0,100,10)
angle
```

```python
V2=np.array([1.092,1.009,0.965,0.895,0.794,0.669,0.523,0.355,0.177,-0.013])
V2
```

```python
k1,k2,k3=np.polyfit(angle,V2,2)
xs=np.linspace(-5,95,300)
ys=k1*xs**2+k2*xs+k3
plt.plot(angle,V2,'o',label='Experiment Data')
plt.plot(xs,ys,label='Fit')
plt.plot(xs,0*xs,'green',0*ys,ys,'green')
plt.xlabel('Angle/degree',fontsize=14)
plt.ylabel('Output Voltage/V',fontsize=14)
plt.legend(frameon=False)
```

```python
Bx=np.array([1.423,1.469,1.492,1.499,1.499,1.499,1.499,1.498,1.489,1.468,1.429])
ratio=Bx/1.499
x=np.arange(-0.5,0.6,0.1)
plt.figure(figsize=(10,8))
plt.plot(x,ratio,'o',label='Expereiment Data')
xs=np.linspace(-0.55,0.55,100)
ys=5**1.5/16*(1/(1+(0.5+xs)**2)**1.5+1/(1+(0.5-xs)**2)**1.5)
plt.plot(xs,ys,label='Theoretical Prediction')
plt.xlabel('Relative Position $\\frac{x}{R}$',fontsize=18)
plt.ylabel('Relative Magnetic Induction $\\frac{B(x)}{B(0)}$',fontsize=18)
plt.legend(frameon=False)
```
