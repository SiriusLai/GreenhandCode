```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
x=np.array([2,2.47712,2.77815,3,3.47712,3.77815,4,4.47712,4.477815,5,5.47712,5.77815,6,6.47712,6.77815,7])
y=np.array([41,127,274,560,923,296,165,53,26.6,15.9,5.31,2.66,1.59,0.53,0.26,0.16])*10**-4
f=interpolate.CubicSpline(x,y,bc_type='natural')
sx=np.linspace(1,7,100)
sy=f(sx)
plt.plot(x,y,'o',sx,sy)
```
