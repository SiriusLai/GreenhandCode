```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
ni=1
nt=1.5
i=np.linspace(0,np.pi/2,100)
t=np.arcsin(ni*np.sin(i)/nt)
r_perpen=-np.sin(i-t)/np.sin(i+t)
r_parall=np.tan(i-t)/np.tan(i+t)
t_perpen=2*np.sin(t)*np.cos(i)/np.sin(i+t)
t_parall=2*np.sin(t)*np.cos(i)/np.sin(i+t)/np.cos(i-t)
fig=plt.figure(figsize=(10,8))
plt.plot(i,r_perpen,label='$r_{\\bot}$')
plt.plot(i,r_parall,label='$r_{//}$')
plt.plot(i,t_perpen,label='$t_{\\bot}$')
plt.plot(i,t_parall,label='$t_{//}$')
plt.plot(i,0*i,'black')
plt.legend(frameon=False,fontsize=18)
```
